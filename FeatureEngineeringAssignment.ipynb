1. What is a parameter?  
Ans : A parameter is a value inside a model that controls how the model behaves. When the model learns, these parameters change. Examples include weights in linear regression or coefficients in a neural network.

2. What is correlation?  
Ans: Correlation is a statistical measure that indicates how strongly two variables move together and the direction of their relationship.


3. What does negative correlation mean?  
Ans: Negative correlation refers to a relationship where one variable increases while the other decreases, showing opposite directional movement.

4. Define Machine Learning. What are the main components in Machine Learning?  
Ans: Machine Learning is a method through which systems learn patterns from data and make predictions or decisions.
Main components include:  Data, Model, Loss function, Optimizer or learning algorithm

5. How does loss value help in determining whether the model is good or not?  
Ans: Loss value indicates how far model predictions deviate from actual values. A lower loss value indicates a more accurate and well-performing model.

6. What are continuous and categorical variables?  
Ans: Continuous variables: numerical values that can take any value within a range.  
Categorical variables: values grouped into categories such as labels or classes.


7. How do we handle categorical variables in Machine Learning? What are the common techniques?  
Ans: Categorical variables are converted into numerical form for model usage. Common techniques include:
Label Encoding, One Hot Encoding, Ordinal Encoding



8. What do you mean by training and testing a dataset?  
Ans: Training data is used to teach the model patterns and relationships.
Testing data is used to evaluate how well the trained model performs on unseen data.

17. What is causation? Explain difference between correlation and causation with an example.  
Ans: Causation means that one variable directly affects or produces a change in another variable.
Correlation only indicates that two variables move together, not that one causes the other.
Example: Ice-cream sales and drowning incidents both increase during summer (correlation), but ice-cream sales do not cause drowning (not causation).



18. What is an Optimizer? What are different types of optimizers? Explain each with an example.  
Ans: An optimizer is an algorithm that updates model parameters to reduce loss during training.
Examples include:

SGD (Stochastic Gradient Descent): updates parameters using gradient direction

Adam: adaptive learning method combining momentum and RMSProp

RMSProp: updates parameters using a moving average of squared gradients
