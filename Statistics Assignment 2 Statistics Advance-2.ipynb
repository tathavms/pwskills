1: What is hypothesis testing in statistics?  
Ans: Hypothesis testing is a statistical method used to make decisions about a population based on sample data. 
It begins with a claim, called a hypothesis, and uses sample information to determine whether there 
is enough evidence to support or reject that claim. 
It helps identify whether an observed effect is real or due to random chance.

2: What is the null hypothesis, and how does it differ from the alternative hypothesis?  
Ans: The null hypothesis states that there is no effect or no difference. It assumes that a population parameter, such as the mean, is equal to a specific value.
The alternative hypothesis states that there is an effect or a difference. It claims that the parameter is not equal, greater than, or less than the value in the null hypothesis.
Hypothesis testing examines data to see whether the null hypothesis should be rejected in favor of the alternative hypothesis.

3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.  
Ans: The significance level, written as α (alpha), is the probability of rejecting the null hypothesis when it is actually true. Common choices are 0.05 or 0.01.  
After calculating the test statistic and p-value, the p-value is compared with α:  
If p-value < α, the null hypothesis is rejected.  
If p-value ≥ α, the null hypothesis is not rejected.  
The significance level controls how strict the decision rule is in hypothesis testing.

4: What are Type I and Type II errors? Give examples of each.  
Ans: A Type I error occurs when the null hypothesis is rejected even though it is true. Example: concluding that a new medicine works better than the old one when there is actually no real difference.  
A Type II error occurs when the null hypothesis is not rejected even though it is false. Example: concluding that a new teaching method is not better than the old one when it actually improves student performance.

5: What is the difference between a Z-test and a T-test? Explain when to use each.  
Ans: A Z-test is used when the population standard deviation is known and the sample size is large (usually n ≥ 30). In this case, the sampling distribution follows the standard normal distribution.  
A T-test is used when the population standard deviation is unknown and the sample size is small. The test statistic follows the t-distribution with (n − 1) degrees of freedom.  
A Z-test is used for large samples with known variance, while a T-test is used for small samples or when the variance is unknown.

6: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
p = 0.5
size = 1000

x = np.random.binomial(n, p, size)

plt.hist(x, bins=range(0, n+2), align='left', rwidth=0.8)
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution n=10, p=0.5')
plt.show()
```

8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from scipy.stats import norm

np.random.seed(0)
x = np.random.normal(100, 15, 1000)

mean = np.mean(x)
std = np.std(x, ddof=1)
se = std / sqrt(len(x))

z = norm.ppf(0.975)
ci_lower = mean - z * se
ci_upper = mean + z * se

print(mean)
print(std)
print(ci_lower)
print(ci_upper)

plt.hist(x, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simulated Normal Data')
plt.show()
```

9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

def z_scores(data):
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    return (data - mean) / std

sample_data = [10, 12, 9, 15, 14, 11, 13, 12, 16, 10]

z = z_scores(sample_data)

print(z)

plt.hist(z, bins=5)
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.title('Histogram of Z-scores')
plt.show()
```
