Question 1: Compare and contrast NLTK and spaCy in terms of features, ease of use, and performance.

Answer:

Features: NLTK (Natural Language Toolkit) is a comprehensive library built mainly for academia and research. It provides access to over 50 corpora and lexical resources, offering many redundant algorithms for a single task (e.g., multiple stemmers). spaCy, on the other hand, is built strictly for production and industrial use cases. It features pre-trained, state-of-the-art transformer pipelines and provides exactly one optimized algorithm for each task.  

Ease of Use: NLTK treats text as a string or list of strings, requiring us to manually chain tokenization, tagging, and parsing steps together. spaCy takes an object-oriented approach. When we pass text through its pipeline, it automatically performs tokenization, parsing, and named entity recognition all at once, returning a fully processed Doc object.

Performance: spaCy is significantly faster than NLTK because its core algorithms are written in heavily optimized Cython. While NLTK is excellent for breaking down and experimenting with individual linguistic sub-steps, spaCy is much better suited for handling large-scale real-world datasets efficiently.

Question 2: What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?  

Answer:
TextBlob is a beginner-friendly Python library built on top of NLTK and another pattern-matching library. It acts as an easy-to-use wrapper, hiding complex linguistic workflows behind a very simple, intuitive API.

Sentiment Analysis: TextBlob simplifies this by providing a .sentiment property on any text object. We do not need to train a model or manually extract features. It instantly returns two simple metrics: Polarity (ranging from -1.0 for negative to +1.0 for positive) and Subjectivity (ranging from 0.0 for objective fact to 1.0 for personal opinion).  

Translation: It allows us to translate or detect languages using a simple .translate() or .detect_language() method call. Under the hood, it handles the complex requests required to hook into translation engines, enabling users to translate multi-language text with just a single line of code.

Question 3: Explain the role of Stanford NLP in academic and industry NLP Projects.  

Answer:
Stanford NLP bridges the gap between deep academic research and robust industry implementations.

Academic Role: Developed by the Stanford Natural Language Processing Group, it serves as a baseline standard for core linguistic analysis. It offers highly accurate grammatical parsing, dependency parsing, and coreference resolution. Academics rely on it heavily because it is deeply rooted in statistical linguistics and provides state-of-the-art accuracy across dozens of human languages.  

Industry Role: In commercial applications, Stanford NLP is heavily utilized whenever high-precision grammatical structure analysis is required (such as extract-transform-load pipelines or legal text analysis). While it can be more computationally demanding than production-first frameworks like spaCy, its industry models are highly accurate, multi-lingual, and often wrapped inside other production systems to generate dependable text features.

Question 4: Describe the architecture and functioning of a Recurrent Neural Network (RNN).  

Answer:
A Recurrent Neural Network (RNN) is a type of artificial neural network designed specifically to process sequential data, such as time series or natural language text.  

Architecture: Unlike traditional feedforward neural networks that process inputs independently, an RNN contains a recurrent loop. This loop allows information to persist. Structurally, it consists of an input layer, a hidden layer containing a hidden state (internal memory), and an output layer. When visualized across time, it can be "unrolled" into a chain of repeating neural network modules.

Functioning:   
1. The network processes text token by token (or word by word).
2. At any given step t, the hidden layer takes two inputs: the current input vector xtвЂ‹ and the previous hidden state htв€’1вЂ‹.
3. It combines these inputs using internal weights and an activation function (typically tanh) to calculate the new hidden state htвЂ‹.
4. This new hidden state acts as the network's updated memory, which is then passed forward to the next step (t+1) and used to generate the current step's output ytвЂ‹.

Question 5: What is the key difference between LSTM and GRU networks in NLP applications?  

Answer:
The key structural difference between an LSTM (Long Short-Term Memory) and a GRU (Gated Recurrent Unit) network lies in their internal gating mechanisms and how they manage their internal memory.

LSTM: Uses three gates (Forget Gate, Input Gate, Output Gate) and splits its memory into two distinct states: the Cell State (ctвЂ‹, which acts as a long-term memory belt) and the Hidden State (htвЂ‹, which acts as short-term memory).

GRU: Offers a simplified architecture by merging the cell state and hidden state into a single hidden state (htвЂ‹). It uses only two gates: an Update Gate (which determines how much past information to keep) and a Reset Gate (which determines how much past information to forget).

Impact on NLP: Because GRUs have fewer parameters, they train significantly faster and require less data to converge, making them ideal for smaller datasets or resource-constrained environments. However, LSTMs can sometimes capture longer dependencies better when trained on large, highly complex textual data.

Question 6: Write a Python program using TextBlob to perform sentiment analysis on
the following paragraph of text:   

вЂњI had a great experience using the new mobile banking app. The interface is intuitive,
and customer support was quick to resolve my issue. However, the app did crash once
during a transaction, which was frustrating"   

Your program should print out the polarity and subjectivity scores.

```python
from textblob import TextBlob

text = ("I had a great experience using the new mobile banking app. The interface is intuitive, "
        "and customer support was quick to resolve my issue. However, the app did crash once "
        "during a transaction, which was frustrating.")

blob = TextBlob(text)
print(f"Polarity Score:     {blob.sentiment.polarity}")
print(f"Subjectivity Score: {blob.sentiment.subjectivity}")
```

Question 7: Given the sample paragraph below, perform string tokenization and
frequency distribution using Python and NLTK:  
вЂњNatural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.вЂќ


```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

nltk.download('punkt')
text = ("Natural Language Processing (NLP) is a fascinating field that combines linguistics, "
        "computer science, and artificial intelligence. It enables machines to understand, "
        "interpret, and generate human language. Applications of NLP include chatbots, "
        "sentiment analysis, and machine translation. As technology advances, the role of NLP "
        "in modern solutions is becoming increasingly critical.")

tokens = word_tokenize(text.lower())
freq_dist = FreqDist(tokens)
print("TOKENS ---")
print(tokens)
print("\n TOP 5 Token FREQUENCIES ---")
for word, frequency in freq_dist.most_common(5):
    print(f"'{word}': {frequency}")
```

Question 8: Implement a basic LSTM model in Keras for a text classification task using
the following dummy dataset. Your model should classify sentences as either positive
(1) or negative (0).    
Dataset  
texts = [  
вЂњI love this projectвЂќ, #Positive  
вЂњThis is an amazing experienceвЂќ, #Positive  
вЂњI hate waiting in lineвЂќ, #Negative  
вЂњThis is the worst serviceвЂќ, #Negative  
вЂњAbsolutely fantastic!вЂќ #Positive  
]  
labels = [1, 1, 0, 0, 1]  
Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on  
this data. You may use Keras with TensorFlow backend.

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

texts = [
    "I love this project", 
    "This is an amazing experience", 
    "I hate waiting in line", 
    "This is the worst service", 
    "Absolutely fantastic!"
]
labels = np.array([1, 1, 0, 0, 1])

tokenizer = Tokenizer(num_words=100)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
max_length = max(len(s) for s in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length)
model = Sequential([
    Embedding(input_dim=100, output_dim=8, input_length=max_length),
    LSTM(units=8),
    Dense(units=1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(padded_sequences, labels, epochs=5, verbose=1)

print("\nModel training complete")
```

Question 9: Using spaCy, build a simple NLP pipeline that includes tokenization,
lemmatization, and entity recognition. Use the following paragraph as your dataset:  

вЂњHomi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of IndiaвЂ™s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India.вЂќ  

Write a Python program that processes this text using spaCy, then prints tokens, their
lemmas, and any named entities found.


```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = ("Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the "
        "development of India's atomic energy program. He was the founding director of the Tata "
        "Institute of Fundamental Research (TIFR) and was instrumental in establishing the "
        "Atomic Energy Commission of India.")

doc = nlp(text)

print("TOKENS AND LEMMAS (First 10 words) ---")
for token in list(doc)[:10]:
    print(f"Token: {token.text:<12} -> Lemma: {token.lemma_}")

print("\nNAMED ENTITIES FOUND ---")
for ent in doc.ents:
    print(f"Entity: {ent.text:<40} | Label: {ent.label_}")
```
