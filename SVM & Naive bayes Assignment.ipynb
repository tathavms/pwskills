1. What is a Support Vector Machine (SVM)?   
Ans: A Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. It works by finding the optimal hyperplane that separates data points of different classes with the maximum possible margin.

2. What is the difference between Hard Margin and Soft Margin SVM? 
Ans: 
* Hard Margin SVM: Assumes the data is perfectly linearly separable and allows no misclassifications, meaning every point must be outside the margin.  
* Soft Margin SVM: Uses a slack variable to allow some misclassifications or points within the margin, making it robust to noise and non-linearly separable data.

3. What is the mathematical intuition behind SVM?   
Ans: The intuition is to solve a constrained optimization problem, maximize the margin while ensuring all data points are correctly classified according to the constraint .

4. What is the role of Lagrange Multipliers in SVM?   
Ans: Lagrange Multipliers are used to convert the constrained optimization problem into a dual form. This allows the algorithm to focus on the support vectors rather than all data points and facilitates the use of the "Kernel Trick".

5. What are Support Vectors in SVM?   
Ans: Support vectors are the data points that lie closest to the decision boundary (hyperplane). They are the critical elements of the training set because if they were removed, the position of the dividing hyperplane would change.

6. What is a Support Vector Classifier (SVC)?   
Ans: SVC is the application of SVM specifically for classification problems, where the goal is to predict discrete class labels by finding the best separating hyperplane.

7. What is a Support Vector Regressor (SVR)?   
Ans: SVR is an adaptation of SVM for regression tasks. Instead of separating classes, it tries to fit the best line within a predefined error threshold, aiming to keep as many points as possible within a "tube" around the line.

8. What is the Kernel Trick in SVM?   
Ans: The Kernel Trick is a mathematical method that allows SVM to operate in a high-dimensional feature space without explicitly calculating the coordinates of the data in that space. It enables the model to learn non-linear decision boundaries.

9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.   
Ans: 
* Linear Kernel: Best for linearly separable data, it is fast and has fewer parameters.  
* Polynomial Kernel: Represents the similarity of vectors in a feature space over polynomials of the original variables, allowing for curved boundaries.  
* RBF (Radial Basis Function) Kernel: The most common kernel, it can map data into an infinite-dimensional space and handle highly complex non-linear relationships.

10. What is the effect of the C parameter in SVM?   
Ans: 
* Small C: Makes the margin larger even if it misclassifies more points (focuses on simplicity/generalization).  
* Large C: Makes the margin smaller to ensure more training points are classified correctly (focuses on accuracy, but risks overfitting).

11. What is the role of the Gamma parameter in RBF Kernel SVM?   
Ans: Gamma defines how far the influence of a single training example reaches. High gamma means only nearby points are considered (leading to a complex, tight boundary), while low gamma means far-away points are also considered (leading to a smoother boundary).

12. What is the Naive Bayes classifier, and why is it called "Naive"?   
Ans: It is a probabilistic classifier based on Bayes' Theorem. It is called "naive" because it assumes that all features are completely independent of each other given the class label, which is rarely true in real-world data.

13. What is Bayes' Theorem?   
Ans: Bayes' Theorem provides a way to calculate the probability of a hypothesis(H) given prior knowledge or evidence(E): P(H∣E)=P(E)P(E∣H)⋅P(H)​


14. Explain the differences between Gaussian Naive Bayes, Multinomial Naive Bayes, and Bernoulli Naive Bayes.  
Ans: 
* Gaussian: Used when features follow a normal (Gaussian) distribution (e.g., height, weight).
* Multinomial: Used for discrete counts (e.g., word frequency in text classification).
* Bernoulli: Used when features are binary (e.g., word presence or absence).

15. When should you use Gaussian Naive Bayes over other variants?   
Ans: You should use Gaussian Naive Bayes when your continuous input features are expected to follow a bell-curve distribution.

16. What are the key assumptions made by Naive Bayes?   
Ans: The primary assumption is Conditional Independence, every feature is assumed to contribute independently to the probability of the class.

17. What are the advantages and disadvantages of Naive Bayes?   
Ans: 
* Advantages: Very fast, works well with high-dimensional data, and requires less training data.
* Disadvantages: The independence assumption is often violated, leading to poor probability estimates.

18. Why is Naive Bayes a good choice for text classification?   
Ans: It is highly efficient at handling large vocabularies (high-dimensional feature spaces) and works surprisingly well even with the independence assumption in word counts.

19. Compare SVM and Naive Bayes for classification tasks.   
Ans: SVM is generally more accurate for complex datasets and resistant to overfitting in high dimensions, while Naive Bayes is much faster and simpler to implement for massive, categorical datasets like text.

20. How does Laplace Smoothing help in Naive Bayes?   
Ans: It prevents the zero-probability problem. If a word appears in the test set but not the training set, the probability would be zero without smoothing, which would make the entire prediction zero.

24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

iris = datasets.load_iris()
X = iris.data[:, :2] 
y = iris.target

poly_svc = SVC(kernel='poly', degree=3, C=1.0)
poly_svc.fit(X, y)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

Z = poly_svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('SVM with Polynomial Kernel (Degree 3)')
plt.show()
```

27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

iris = datasets.load_iris()
X = iris.data[:, :2] 
y = iris.target

c_values = [0.1, 1, 100]
plt.figure(figsize=(15, 4))

for i, C in enumerate(c_values):
    clf = SVC(kernel='linear', C=C).fit(X, y)
    
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.subplot(1, 3, i + 1)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.title(f'SVM Decision Boundary (C={C})')

plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'linear', 'poly']
}

grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=0, cv=5)

grid.fit(X_train, y_train)

print(f"Best Parameters: {grid.best_params_}")

grid_predictions = grid.predict(X_test)
print(f"Accuracy with Tuned Parameters: {accuracy_score(y_test, grid_predictions) * 100:.2f}%")
```


32. Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and check it improve accuracy. 


```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

#Creating a synthetic imbalanced dataset (90% class 0, 10% class 1)
X, y = make_classification(n_samples=1000, n_features=10, n_clusters_per_class=1,
                           weights=[0.9, 0.1], flip_y=0, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#without class weighting
model_standard = SVC(kernel='linear')
model_standard.fit(X_train, y_train)
y_pred_standard = model_standard.predict(X_test)

# with class weighting ('balanced')
model_weighted = SVC(kernel='linear', class_weight='balanced')
model_weighted.fit(X_train, y_train)
y_pred_weighted = model_weighted.predict(X_test)

print("--- Standard SVM (No Weights) ---")
print(classification_report(y_test, y_pred_standard))

print("\n--- Weighted SVM (class_weight='balanced') ---")
print(classification_report(y_test, y_pred_weighted))
```


35. Write a Python program to perform feature selection before training a Naive Bayes classifier and compare results. 

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model_all = GaussianNB()
model_all.fit(X_train, y_train)
acc_all = accuracy_score(y_test, model_all.predict(X_test))

selector = SelectKBest(score_func=f_classif, k=5)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

model_selected = GaussianNB()
model_selected.fit(X_train_selected, y_train)
acc_selected = accuracy_score(y_test, model_selected.predict(X_test_selected))

print(f"Accuracy with all 30 features: {acc_all * 100:.2f}%")
print(f"Accuracy with top 5 features:  {acc_selected * 100:.2f}%")
```

36. Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and One-vs-One (OvO) strategies on the Wine dataset and compare their accuracy. 

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

wine = load_wine()
X, y = wine.data, wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# One-vs-Rest (OvR) strategy
ovr_model = OneVsRestClassifier(SVC(kernel='linear'))
ovr_model.fit(X_train_scaled, y_train)
ovr_acc = accuracy_score(y_test, ovr_model.predict(X_test_scaled))

# One-vs-One (OvO) strategy
ovo_model = OneVsOneClassifier(SVC(kernel='linear'))
ovo_model.fit(X_train_scaled, y_train)
ovo_acc = accuracy_score(y_test, ovo_model.predict(X_test_scaled))

print(f"One-vs-Rest (OvR) Accuracy: {ovr_acc * 100:.2f}%")
print(f"One-vs-One  (OvO) Accuracy: {ovo_acc * 100:.2f}%")
```

38. Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy.

```python
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
import numpy as np

data = load_iris()
X, y = data.data, data.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = SVC(kernel='linear')

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(model, X_scaled, y, cv=skf)

print(f"Individual Fold Accuracies: {scores}")
print(f"Average Accuracy: {np.mean(scores) * 100:.2f}%")
```


40. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy. 


```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model_all = SVC(kernel='linear')
model_all.fit(X_train_scaled, y_train)
acc_all = accuracy_score(y_test, model_all.predict(X_test_scaled))

svc = SVC(kernel="linear")
selector = RFE(estimator=svc, n_features_to_select=5, step=1)
X_train_rfe = selector.fit_transform(X_train_scaled, y_train)
X_test_rfe = selector.transform(X_test_scaled)

model_rfe = SVC(kernel='linear')
model_rfe.fit(X_train_rfe, y_train)
acc_rfe = accuracy_score(y_test, model_rfe.predict(X_test_rfe))

print(f"Accuracy with all 30 features: {acc_all * 100:.2f}%")
print(f"Accuracy with top 5 RFE features: {acc_rfe * 100:.2f}%")
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler

data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print("Detailed Classification Report:")
print(classification_report(y_test, y_pred))

#individually:
p = precision_score(y_test, y_pred)
r = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Individual Metrics for the Positive Class:")
print(f"Precision: {p:.4f}")
print(f"Recall:    {r:.4f}")
print(f"F1-Score:  {f1:.4f}")
```

43. Write a Python program to train an SVM Classifier and visualize the Confusion Matrix using seaborn. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix

iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = SVC(kernel='linear')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=iris.target_names, 
            yticklabels=iris.target_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('SVM Confusion Matrix - Iris Dataset')
plt.show()
```

46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, PrecisionRecallDisplay
from sklearn.preprocessing import StandardScaler

data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)

y_score = model.decision_function(X_test_scaled)
precision, recall, _ = precision_recall_curve(y_test, y_score)

plt.figure(figsize=(8, 6))
display = PrecisionRecallDisplay(precision=precision, recall=recall)
display.plot()
plt.title('Precision-Recall Curve (SVM)')
plt.grid(True)
plt.show()
```
