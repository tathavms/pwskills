1.	What is a random variable in probability theory?  
Ans: A random variable is a quantity that takes different values depending on the result of a random experiment. It connects real-life random events to numerical values that can be analyzed mathematically.

2.	What are the types of random variables?  
Ans: There are two main types of random variables: discrete and continuous. A discrete random variable takes countable values like 0, 1, 2, etc. A continuous random variable takes any value within a range, such as height, weight, or temperature.

3.	Explain the difference between discrete and continuous distributions.  
Ans: A discrete distribution represents data that can take only specific, separate values. For example, the number of students in a class. A continuous distribution represents data that can take any value within an interval, like time or distance

4.	What is a binomial distribution, and how is it used in probability?  
Ans: A binomial distribution shows the number of successes in a fixed number of independent trials, each with the same probability of success. It is used to model events like coin tosses or the number of defective products in a batch.

5.	What is the standard normal distribution, and why is it important?  
Ans: The standard normal distribution is a type of normal distribution with a mean of 0 and a standard deviation of 1. It is important because it helps in calculating probabilities and comparing different normal distributions using z-scores.

6.	What is the Central Limit Theorem (CLT), and why is it critical in statistics?  
Ans: The Central Limit Theorem states that when the sample size is large enough, the sampling distribution of the sample mean becomes approximately normal, no matter what the population distribution is. It is critical because it allows the use of normal probability methods for analyzing sample data.

7.	What is the significance of confidence intervals in statistical analysis?  
Ans: Confidence intervals give a range within which the true population parameter is expected to lie. They show how reliable the sample estimate is and help in decision-making based on sample data.

8.	What is the concept of expected value in a probability distribution?  
Ans: The expected value is the average value that a random variable takes in the long run. It represents the weighted average of all possible outcomes, each multiplied by its probability.

9.	Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np 
import matplotlib.pyplot as plt 
x = np.random.normal(50, 5, 1000) 
print(np.mean(x)) 
print(np.std(x)) 
plt.hist(x, bins=30) 
plt.show()
```

10.	You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]   
● Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.   
● Write the Python code to compute the mean sales and its confidence interval.

```python
import numpy as np 
import scipy.stats as st 
daily_sales = [220,245,210,265,230,250,260,275,240,255,235,260,245,250,225,270,265,255,250,260] 
mean = np.mean(daily_sales) 
std = np.std(daily_sales, ddof=1) 
ci = st.t.interval(0.95, len(daily_sales)-1, loc=mean, scale=std/np.sqrt(len(daily_sales))) 
print(mean) 
print(ci)
```
